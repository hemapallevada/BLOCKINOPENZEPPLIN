```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori
```

```
url='https://raw.githubusercontent.com/ArunkumarGoge/DataAnalyticsLab/master/Exercise2_Practise.csv'
name=[i for i in range(0,20)]
df = pd.read_csv(url, names=name,header=None)
```

```
print(df.shape)
```

```
print(df.head())
```

```
df.fillna(0,inplace=True)
print(df.head())
```

```
transactions = []
for i in range(0,len(df)):
    transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```
rules = apriori(transactions,min_support=0.00003,min_confidance=0.00002,min_lift=3,max_length=2)
listaas=list(rules)
mx=0.0
print(listaas)
item=""
for i in range(0,len(listaas)):
  if(mx<listaas[i][1]):
    mx=listaas[i][1]
    item=listaas[i][0]
print(item)
#max_support=max(listaas[:][1])
```
